```python
# проводим стандартизацию данных
sample_s = (np.log10(df_new.width) - np.log10(df_new.width).mean()/np.log10(df_new.width).std()) 
print(sample_s)
hist_s = plt.hist(sample_s, bins=10, density=True)
```

```python
# провожу разные тесты
h0_1_p_value = stats.shapiro(sample_s).pvalue
h0_1_p_value_1 = stats.shapiro(np.log10(df_new.width)).pvalue
h0_1_p_value_2 = stats.kstest(sample_s, cdf=stats.norm.cdf).pvalue
h0_1_p_value_3 = stats.normaltest(sample_s).pvalue
h0_1_p_value_4 = stats.ks_2samp(sample_s, stats.norm.rvs(size=len(sample_s))).pvalue
h0_1_p_value_5 = stats.mannwhitneyu(sample_s, stats.norm.rvs(size=len(sample_s))).pvalue

print('p-value:', h0_1_p_value, h0_1_p_value_1, h0_1_p_value_2, h0_1_p_value_3, h0_1_p_value_4, h0_1_p_value_5, sep='\n' )
```

```python
# Процедура для приведения колонки designer к нормальному виду
```

```python
import contextlib

def cleanDesigner(value, removeIKEA=False, emptyValue=np.nan):
    if not isinstance(value, str):
        return value
    if len(value)>0 and value[0].isdigit():
        return emptyValue
    designers = value.split('/')
    if removeIKEA:
        with contextlib.suppress(Exception):
            designers.remove('IKEA of Sweden')
    return '/'.join(sorted(designers)) if len(designers) > 0 else emptyValue
```

```python
# добавим новую колонку designer_clean - поле designer без мусора с отсортированным списком дизайнеров. 
# пустые значения заполним IKEA of Sweden
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2020/2020-11-03/ikea.csv',
                index_col=0).drop_duplicates()
df['designer_clean'] = df['designer'].apply(cleanDesigner, args=(False, 'IKEA of Sweden'))
df['designer_clean'].value_counts()
```

```python
df[['depth', 'height', 'width']] = df[['depth', 'height', 'width']].fillna(df[['depth', 'height', 'width']].mean())
# замена пустых значений на среднее
df.count()
```

```python
X = df[['depth', 'width', 'height', 'category', 'designer_clean', 'other_colors']]
Y = df['price']

X_train, X_test, Y_train, Y_test = sk.model_selection.train_test_split(X, Y, test_size=0.2, random_state=42)

numeric_transf = Pipeline(steps=[
    ('scaler', StandardScaler()),
    ('impute', SimpleImputer(strategy='median'))
])

categorical_transf = Pipeline(steps=[
    ('impute', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])

col_prepr = ColumnTransformer(transformers=[
    ('numeric', numeric_transf, ['depth', 'width', 'height']),
    ('categorical', categorical_transf, ['category', 'designer_clean', 'other_colors'])
])

dtr = Pipeline(steps=[
    ('col_prep', col_prepr),
    ('dtr', DecisionTreeRegressor(max_depth=10, random_state=42))
])

dtr.fit(X_train, Y_train)
dtr_predict = dtr.predict(X_test)

print('R^2 : {:.5f}'.format(dtr.score(X_test, Y_test)))
print('MAE : {:.5f}'.format(sk.metrics.mean_absolute_error(dtr_predict, Y_test)))
print('MSE : {:.5f}'.format(np.sqrt(sk.metrics.mean_absolute_error(dtr_predict, Y_test))))
```

```python
# Получив такие результаты по первой модели ясно что модель нужно обучать для увеличения R^2

# Для улучшения прогноза нужно
# 1. Подобрать другую стратегию заполнения отсутствующих габаритов (заполнено средним по всей выборке)
# 2. Использовать другой Encoder для категорий и дизайнеров
# 3. Подобрать другой регрессор
# 4. Подобрать наиболее оптимальные параметры регрессора

# 1. В первоначальной модели мы заполняли отсутствующие габариты средним значением по всей выборке.
# Более корректно заполнить средними значениями по категории
# 2. Закодировать категории и дизайнеров средней (медиана) ценой по категории.

# Добавляю в дата фрейм новые поля
```

```python
df['other_colors_1'] = df['other_colors'].map(dict(Yes=1, No=0))

median_d = df.groupby(['category'])['depth'].median()
median_h = df.groupby(['category'])['height'].median()
median_w = df.groupby(['category'])['width'].median()

median_price = df.groupby(['category'])['price'].median()
median_dsgn = df.groupby(['designer_clean'])['price'].median()

df = df.set_index(['category'])
df['depth_1'] = df['depth'].fillna(median_d)
df['height_1'] = df['height'].fillna(median_h)
df['width_1'] = df['width'].fillna(median_w)
df['category_median_price'] = median_price
df=df.reset_index()

df = df.set_index(['designer_clean'])
df['designer_median_price'] = median_dsgn
df=df.reset_index()

df.head()
```

```python
# Процедура, которая будет принимать на вход выборки и 
# рассчитывать оценки для разных регрессов
```

```python
def getBestRegressor(X, Y):
    X_train, X_test, Y_train, Y_test = sk.model_selection.train_test_split(X, Y, test_size=0.2, random_state=42)
    # модели из которых будет делаться перебор
    models = [
        sk.linear_model.LinearRegression(),
        sk.linear_model.LassoCV(),
        sk.linear_model.RidgeCV(),
        sk.svm.SVR(kernel='linear'),
        sk.neighbors.KNeighborsRegressor(n_neighbors=16),
        sk.tree.DecisionTreeRegressor(max_depth=10, random_state=42),
        RandomForestRegressor(random_state=42),
        GradientBoostingRegressor()
    ]
    
    TestModels = pd.DataFrame()
    tmp = {}
    # для каждой модели из списка
    for model in models:
        # получают имя модели
        m = str(model)
        tmp['Model'] = m[:m.index('(')]
        # обучаем модель
        model.fit(X_train, Y_train)
        # вычисляем R^2 - коэфф детерминации
        tmp['R^2'] = '{:.5}'.format(model.score(X_test, Y_test))
        # вычисляем МАЕ - средний модуль отклонения
        tmp['MAE'] = '{:.5}'.format(sk.metrics.mean_absolute_error(model.predict(X_test), Y_test))
        # вычисляем RMSE - корень из среднего квадрата отклонения
        tmp['RMSE'] = '{:.5}'.format(np.sqrt(sk.metrics.mean_absolute_error(model.predict(X_test), Y_test)))
        
        # записываем данные в итоговый ДатаФрейм
        TestModels = pd.concat([TestModels, pd.DataFrame([tmp])])
        
    # делаем индекс по названию модели
    TestModels.set_index('Model', inplace=True)
    return{'model': TestModels, 'X_train': X_train, 'Y_train': Y_train, 'X_test': X_test, 'Y_test': Y_test}
                        
```

```python
# ТЕСТ № 1
# Беру габариты рассчитанные по медиане в разрезе категорий, категории и дизайнеры закодированные медианной цены
# Проверяю какой регрессор покажет лучший результат и как изменится оценка для модели на основе дерева решений
```

```python
X1 = df[['width_1', 'depth_1', 'height_1', 'category_median_price', 'designer_median_price', 'other_colors_1']]
Y1 = df['price']

test1 = getBestRegressor(X1, Y1)
test1['model'].sort_values(by='R^2', ascending=False)
```

```python
df3 = df.dropna(subset=['width', 'height', 'depth'])
X3 = df3[['width', 'height', 'depth', 'category_median_price', 'designer_median_price2', 'other_colors_1']]
Y3 = df3['price']

test3 = getBestRegressor(X3, Y3)
test3['model'].sort_values(by='R^2', ascending=False)
```

```python
# Следующий шаг оптимизация гиперпараметров (подставляем самый высокий R^2 по предыдущим тестам)
```

```python
from sklearn.model_selection import GridSearchCV

X_train, X_test, Y_train, Y_test = sk.model_selection.train_test_split(X3, Y3, test_size=0.2, random_state= 42)
forest_grid = GridSearchCV(RandomForestRegressor(), {'n_estimators': [10,50,100, 110, 120,130,140,150,200,500],
                                                    'max_depth': [10,100,200,None],
                                                    'max_features': ['auto', 'sqrt', 'log2']}, cv=5, n_jobs=-1, verbose=3)

forest_grid.fit(X_train, Y_train)
print('Best Estimator :', forest_grid.best_estimator_)
print('Best Score     :', forest_grid.best_score_)
print('')
print('R^2            :{:.5}'.format(sk.metrics.r2_score(Y_test, forest_grid.predict(X_test))))
print('MAE            :{:.5}'.format(sk.metrics.mean_absolute_error(forest_grid.predict(X_test), Y_test)))
print('RMSE           :{:.5}'.format(np.sqrt(sk.metrics.mean_squared_error(forest_grid.predict(X_test), Y_test))))
print('')
print('Feature importance:')
print('________________________________')
for feat, importance in zip(X_train.columns, forest_grid.best_estimator_.feature_importances_):
      print('{:.5}     {f}'.format(importance, f=feat))
      
sns.set_style('whitegrid')
sns.barplot(y=X_train.columns, x=forest_grid.best_estimator_.feature_importances_)
      
```
